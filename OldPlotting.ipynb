```r
options(warn=-1)
options("getSymbols.warning4.0"=FALSE)
suppressMessages(library(quantmod))
normalize_series <- function(xdat) (xdat / coredata(xdat)[1])
symbol.list <- c('SCHV','SCHG','SCHP')
getSymbols(symbol.list)
date.today <- Sys.Date()
date.range <- paste(date.today-10, "/", sep="")
nSCHG <- (normalize_series(Cl(SCHG[date.range]))-1)*100
nSCHV <- (normalize_series(Cl(SCHV[date.range]))-1)*100
nSCHP <- (normalize_series(Cl(SCHP[date.range]))-1)*100
ymax <- max(c(nSCHG, nSCHV, nSCHP))
ymin <- min(c(nSCHG, nSCHV, nSCHP))
plot.ylim <- range(ymin,ymax)
head(nSCHP)
```

```r
#options(repr.plot.width=12, repr.plot.height=4)
myTheme = chart_theme()
myTheme$col$line.col = "blue"
myTheme$rylab = FALSE
myTheme$lylab = FALSE
myPars = chart_pars()
myPars$cex = 1.2
csch <- chart_Series(nSCHV, theme = myTheme, pars=myPars)
csch <- add_TA(nSCHG, on = 1, col = 'red', lty = 1)
csch <- add_TA(nSCHP, on = 1, col = 'green', lty = 1)
csch
title("Foobar")
legend("left", legend=c("SCHG", "SCHP", "SCHV"), col=c("red", "blue", "darkgreen"),cex=0.8)
```

```r
#par(mar = c(5.1, 4.1, 4.1, 2.1))
par(mar = c(5.1, 4.1, 4.1, 2.1), ylab="foo", xlab="bar")
pl <- plot(nSCHP, col="blue", ylim=plot.ylim)
pl <- lines(nSCHG, col="darkgreen")
pl <- lines(nSCHV, col="red")
pl
title("Schwab Indices")
legend("center", legend=c("SCHP"), col=c("blue"), lty=1)
```
