```r
symbol.list <- c('IBM')
to <- as.Date("2005-01-01")
from <- as.Date("1990-01-01")
getSymbols(symbol.list, from=from, to=to)
symbol <- eval(parse(text=symbol.list[1]))
```

```r
date.range <- paste(from, to, sep="/")
options(repr.plot.width=18, repr.plot.height=8)
chartSeries(Cl(symbol[date.range]), theme = chartTheme('white'))
```
