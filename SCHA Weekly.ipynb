```r
options(warn=-1)
options("getSymbols.warning4.0"=FALSE)
suppressMessages(library(quantmod))
suppressMessages(library(lubridate))
```

```r
normalize_series <- function(xdat) xdat / coredata(xdat)[1]
```

```r
getSymbols(c('SCHA', 'SCHV'))
```

```r
target.date <- Sys.Date() 
month(target.date) <- month(target.date) - 6
scha <- SCHA[paste(target.date, "/", sep="")]
schv <- SCHV[paste(target.date, "/", sep="")]
```

```r
mytheme = chart_theme()
mytheme$col$line.col = "green"
mytheme$col$line.lty = 2
chart_Series(normalize_series(Cl(to.weekly(scha)))-1, theme = mytheme,
    TA = c("add_TA(normalize_series(Cl(to.weekly(schv)))-1, on=1, col='red')"))
```
