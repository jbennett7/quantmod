```r
normalize_series <- function(xdat) xdat / coredata(xdat)[1]
```

```r
#symbol.list <- c('^DJI','SCHV','SCHA','SCHD','SHOP')
symbol.list <- c('SCHV','SCHG','SCHP')
```

```r
getSymbols(symbol.list)
```

```r
date.today <- Sys.Date()
date.range <- paste(date.today-30, "/", sep="")
#date.range <- '2018-04-01/2020-05-08'
```

```r
nSCHG <- normalize_series(Cl(SCHG[date.range]))-1
nSCHV <- normalize_series(Cl(SCHV[date.range]))-1
nSCHP <- normalize_series(Cl(SCHP[date.range]))-1
ymax <- max(c(nSCHG, nSCHV, nSCHP))
ymin <- min(c(nSCHG, nSCHV, nSCHP))
plot.ylim <- range(ymin,ymax)
```

```r
options(repr.plot.width=12, repr.plot.height=4)
#par(cex=3)
#par(mfrow=c(1,3))
myTheme = chart_theme()
myTheme$col$line.col = "blue"
myTheme$rylab = FALSE
myTheme$lylab = FALSE
myPars = chart_pars()
myPars$cex = 1.2

csch <- chart_Series(normalize_series(Cl(SCHV[date.range]))-1, theme = myTheme, pars=myPars)
csch <- add_TA(normalize_series(Cl(SCHG[date.range]))-1, on = 1, col = 'red', lty = 1)
csch <- add_TA(normalize_series(Cl(SCHP[date.range]))-1, on = 1, col = 'green', lty = 1)
#title("Foobar", sub = "bingbang")
#csch$Env$actions[[4]] <- NULL
#chart_Series(normalize_series(Cl(SCHG[date.range]))-1, theme = myTheme, pars=myPars,
#    TA = c("add_TA(normalize_series(Cl(SHOP[date.range]))-1, on = 1, col = 'red', lty = 1)"))
#chart_Series(normalize_series(Cl(SCHV[date.range]))-1, theme = myTheme, pars=myPars,
#    TA = c("add_TA(normalize_series(Cl(SHOP[date.range]))-1, on = 1, col = 'red', lty = 1)"))
#csch$Env$name <- "FooOOObar"
#text.exp <- c(expression(text(1 - 1/3, 0.5, name, font = 2, 
#        col = "#444444", offset = 0, cex = 1.1, pos = 4)), 
#                  expression(text(NROW(xdata[xsubset]), 
#        0.5, paste(start(xdata[xsubset]), end(xdata[xsubset]), 
#            sep = " / "), col = 1, adj = c(0, 0), pos = 2)))
#csch$add(text.exp, expr=TRUE)
```
