```r
getSymbols(c('SCHV', 'SCHA', 'SCHO', 'SCHE'))
par(mfrow=c(2,2))
options(repr.plot.width=16, repr.plot.height=12)
mytheme = chart_theme()
myPars <- chart_pars()
myPars$cex=1.2
myPars$mar <- c(3,2,1,1)

date.range <- YY2013_2014
nSCHV <- normalize_series(Cl(SCHV[date.range]))-1
nSCHA <- normalize_series(Cl(SCHA[date.range]))-1
nSCHO <- normalize_series(Cl(SCHO[date.range]))-1
nSCHE <- normalize_series(Cl(SCHE[date.range]))-1
{
chart_Series(
    nSCHO,
    pars=myPars,
    theme=mytheme,
    col='darkorange'
)
add_TA(nSCHV, on=1, col='darkblue')
add_TA(nSCHA, on=1, col='darkgreen')
add_TA(nSCHE, on=1, col='red')
}
title("Y2013-Y2014")
legend("left", legend = c('SCHO', 'SCHV', 'SCHA', 'SCHE'),
       col = c('darkorange', 'darkblue', 'darkgreen', 'red'), lty=1)

date.range <- YY2015_2016
nSCHV <- normalize_series(Cl(SCHV[date.range]))-1
nSCHA <- normalize_series(Cl(SCHA[date.range]))-1
nSCHO <- normalize_series(Cl(SCHO[date.range]))-1
nSCHE <- normalize_series(Cl(SCHE[date.range]))-1
{
chart_Series(
    nSCHO,
    pars=myPars,
    theme=mytheme,
    col='darkorange'
)
add_TA(nSCHV, on=1, col='darkblue')
add_TA(nSCHA, on=1, col='darkgreen')
add_TA(nSCHE, on=1, col='red')
}

title("Y2015-Y2016")
legend("left", legend = c('SCHO', 'SCHV', 'SCHA', 'SCHE'),
       col = c('darkorange', 'darkblue', 'darkgreen', 'red'), lty=1)


date.range <- YY2015_2016
nSCHV <- normalize_series(Cl(SCHV[date.range]))-1
nSCHA <- normalize_series(Cl(SCHA[date.range]))-1
nSCHO <- normalize_series(Cl(SCHO[date.range]))-1
nSCHE <- normalize_series(Cl(SCHE[date.range]))-1
{
chart_Series(
    nSCHO,
    pars=myPars,
    theme=mytheme,
    col='darkorange'
)
add_TA(nSCHV, on=1, col='darkblue')
add_TA(nSCHA, on=1, col='darkgreen')
add_TA(nSCHE, on=1, col='red')
}

title("Y2017-2018")
legend("left", legend = c('SCHO', 'SCHV', 'SCHA', 'SCHE'),
       col = c('darkorange', 'darkblue', 'darkgreen', 'red'), lty=1)


date.range <- YY2019_2020
nSCHV <- normalize_series(Cl(SCHV[date.range]))-1
nSCHA <- normalize_series(Cl(SCHA[date.range]))-1
nSCHO <- normalize_series(Cl(SCHO[date.range]))-1
nSCHE <- normalize_series(Cl(SCHE[date.range]))-1
{
chart_Series(
    nSCHO,
    pars=myPars,
    theme=mytheme,
    col='darkorange'
)
add_TA(nSCHV, on=1, col='darkblue')
add_TA(nSCHA, on=1, col='darkgreen')
add_TA(nSCHE, on=1, col='red')
}
title("Y2019-2020")
legend("left", legend = c('SCHO', 'SCHV', 'SCHA', 'SCHE'),
       col = c('darkorange', 'darkblue', 'darkgreen', 'red'), lty=1)
```
