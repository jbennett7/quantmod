```r
options(warn=-1)
options("getSymbols.warning4.0"=FALSE)
suppressMessages(library(quantmod))
```

```r
normalize_series <- function(xdat) xdat / coredata(xdat)[1]
```

```r
schwab.broad.market <- c('SCHB', 'FNDB')
schwab.dividend <- c('SCHD')
schwab.large.cap <- c('SCHX', 'FNDX')
schwab.growth <- c('SCHG')
schwab.value <- c('SCHV')
schwab.mid.cap <- c('SCHM')
schwab.small.cap <- c('SCHA','FNDA')
schwab.int.large.cap <- c('SCHF', 'FNDF')
schwab.int.small.cap <- c('SCHC', 'FNDC')
schwab.em <- c('SCHE', 'FNDE')
schwab.long.treasuries <- c('SCHQ')
schwab.inter.treasuries <- c('SCHR')
schwab.short.treasuries <- c('SCHO')
schwab.tips <- c('SCHP')
schwab.aggregate.bonds <- c('SCHZ')
schwab.5_10yr.corp.bonds <- c('SCHI')
schwab.1_5yr.corp.bonds <- c('SCHJ')
schwab.reit <- c('SCHH')
schwab.1000 <- c('SCHK')
```

```r
Q1.2018 <- "2018-01-01/2018-03-31"
Q2.2018 <- "2018-04-01/2018-06-30"
Q3.2018 <- "2018-07-01/2018-09-30"
Q4.2018 <- "2018-10-01/2018-12-31"
Q1.2019 <- "2019-01-01/2019-03-31"
Q2.2019 <- "2019-04-01/2019-06-30"
Q3.2019 <- "2019-07-01/2019-09-30"
Q4.2019 <- "2019-10-01/2019-12-31"
Q1.2020 <- "2020-01-01/2020-03-31"
Q2.2020 <- "2020-04-01/2020-06-30"
Q3.2020 <- "2020-07-01/2020-09-30"
Q4.2020 <- "2020-10-01/2020-12-31"
```

```r
getSymbols(schwab.broad.market)
par(mfrow=c(2,2))
options(repr.plot.width=16, repr.plot.height=12)
mytheme = chart_theme()
myPars <- chart_pars()
myPars$cex=1.2
myPars$mar <- c(3,2,1,1)


nSCHB <- normalize_series(Cl(SCHB[Q3.2019]))-1
nFNDB <- normalize_series(Cl(FNDB[Q3.2019]))-1
{
chart_Series(
    nSCHB,
    pars=myPars,
    theme=mytheme,
    col='darkorange'
)
add_TA(nFNDB, on=1, col='darkgreen')
}
title("3rd Q 2019")
legend("left", legend = c('SCHB', 'FNDB'), col = c('darkorange', 'darkgreen'), lty=1)


nSCHB <- normalize_series(Cl(SCHB[Q4.2019]))-1
nFNDB <- normalize_series(Cl(FNDB[Q4.2019]))-1
{
chart_Series(
    nSCHB,
    pars=myPars,
    theme=mytheme,
    col='darkorange'
)
add_TA(nFNDB, on=1, col='darkgreen')
}
title("4th Q 2019")
legend("left", legend = c('SCHB', 'FNDB'), col = c('darkorange', 'darkgreen'), lty=1)

nSCHB <- normalize_series(Cl(SCHB[Q1.2020]))-1
nFNDB <- normalize_series(Cl(FNDB[Q1.2020]))-1
{
chart_Series(
    nSCHB,
    pars=myPars,
    theme=mytheme,
    col='darkorange'
)
add_TA(nFNDB, on=1, col='darkgreen')
}
title("1st Q 2020")
legend("left", legend = c('SCHB', 'FNDB'), col = c('darkorange', 'darkgreen'), lty=1)

nSCHB <- normalize_series(Cl(SCHB[Q2.2020]))-1
nFNDB <- normalize_series(Cl(FNDB[Q2.2020]))-1
{
chart_Series(
    nSCHB,
    pars=myPars,
    theme=mytheme,
    col='darkorange'
)
add_TA(nFNDB, on=1, col='darkgreen')
}
title("2nd Q 2020")
legend("left", legend = c('SCHB', 'FNDB'), col = c('darkorange', 'darkgreen'), lty=1)
```
